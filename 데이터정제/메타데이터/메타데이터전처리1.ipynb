```python
# 진행 상태 알려주는 함수 
# !pip install tqdm
import time
from glob import glob
import os
import glob
import pandas as pd
for i,a in tqdm(enumerate(glob.glob('C:\workspace\eatand_medicinal_plant\\학습용데이터\\*\\*\\*\\*\\*.json'))):
        globals()[f'df{i}'] = pd.read_json(a)
        globals()[f'df{i}'].T
        info        = globals()[f'df{i}'].T.iloc[[0],[2]]
        images      =  globals()[f'df{i}'].T.iloc[[1],7:13]
        annotations =  globals()[f'df{i}'].T.iloc[[2],14:23]
        plants      =  globals()[f'df{i}'].T.iloc[[3],23:31]
        ingredient  =  globals()[f'df{i}'].T.iloc[[4],31:-1]


        abc = [info,images,annotations,plants,ingredient]



        info.rename(index={'info':0},inplace=True)
        images.rename(index={'images':0},inplace=True)
        annotations.rename(index={'annotations':0},inplace=True)
        plants.rename(index={'plants':0},inplace=True)
        ingredient.rename(index={'ingredient':0},inplace=True)

        globals()[f'df{i}'] = pd.concat(abc,axis=1,join='inner',ignore_index=None)
        globals()[f'df{i}'].set_index(globals()[f'df{i}']['wd_plnt_nm'],inplace=True)
        
        
        
        if i == 0:
            sumdf = globals()[f'df{i}']
        else :
            sumdf = pd.concat([sumdf,globals()[f'df{i}']],axis=0)
```
