```python
from glob import glob
import os
import glob
import pandas as pd
for i,a in (enumerate(glob.glob('C:\workspace\eatand_medicinal_plant\\testdata\\220121testdata\\*\\*\\*\\*\\*.json'))):
        globals()[f'df{i}'] = pd.read_json(a)
        globals()[f'df{i}'].T
        info        = globals()[f'df{i}'].T.iloc[[0],[2]]
        images      =  globals()[f'df{i}'].T.iloc[[1],7:13]
        annotations =  globals()[f'df{i}'].T.iloc[[2],14:23]
        plants      =  globals()[f'df{i}'].T.iloc[[3],23:31]
        ingredient  =  globals()[f'df{i}'].T.iloc[[4],31:-1]


        abc = [info,images,annotations,plants,ingredient]



        info.rename(index={'info':0},inplace=True)
        images.rename(index={'images':0},inplace=True)
        annotations.rename(index={'annotations':0},inplace=True)
        plants.rename(index={'plants':0},inplace=True)
        ingredient.rename(index={'ingredient':0},inplace=True)

        globals()[f'df{i}'] = pd.concat(abc,axis=1,join='inner',ignore_index=None)
        globals()[f'df{i}'].set_index(globals()[f'df{i}']['wd_plnt_nm'],inplace=True)
        
        
        
        if i == 0:
            sum_test_df = globals()[f'df{i}']
        else :
            sum_test_df = pd.concat([sum_test_df,globals()[f'df{i}']],axis=0)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train_Allec,X_testec,y_train_Allec,y_testec = train_test_split(
    x_data,y_data_get,test_size=0.2
)
X_trainec,X_valec,y_trainec,y_valec = train_test_split(
    X_train_Allec,y_train_Allec,test_size=0.25
)
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical


mec = Sequential()
mec.add(Dense(150, activation='relu', input_shape=(21,)))
mec.add(Dense(45, activation='relu'))
mec.add(Dense(15, activation='softmax'))

mec.summary()
mec.compile(loss='categorical_crossentropy',
              metrics='accuracy',optimizer='adam')
```

```python
history = mec.fit(X_trainec, y_trainec, epochs=1000,steps_per_epoch=400, validation_data = (X_valec,y_valec), verbose = 1, validation_steps=3)
```
